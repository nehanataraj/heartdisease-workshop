```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from ucimlrepo import fetch_ucirepo
```

## Dataset

```python
# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets
y[y > 0] = 1

# print the data
X
```

```python
# Fill in NaN values with the mean of the column
X = X.fillna(X.mean())
```

## Model

```python
# Create model (the best part)
model = keras.Sequential([
    keras.layers.InputLayer(input_shape=(13,)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.summary()
```

```python
# Compile model (learning rate 0.001)
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Fit model (100 epochs, 32 batch size)
model.fit(X, y, epochs=100, batch_size=32)
```

```python
# Evaluate model
model.evaluate(X, y)
```

## Try it out!

```python
normal = [63,1,3,145,233,1,0,150,0,2.3,0,0,1]
diseased = [67,1,4,160,286,0,0,108,1,1.5,1,3,2]

# Generate predictions (OUTPUT should be [[0], [1]])
model(np.array([normal, diseased]))
```
